# Các thông tin cơ bản của bộ dữ liệu

```python
#Gộp 3 sheet SE Reginal Sales tháng 10, 11, 12 thành 1 file SE Regional Sales.xlsx
sales_data = pd.read_excel('SE Regional Sales.xlsx')
customer_data = pd.read_excel('Customer.xlsx')

# Hiển thị thông tin cơ bản về dữ liệu
print("Thông tin dữ liệu bán hàng:")
print(sales_data.info())

print("\nThông tin dữ liệu khách hàng:")
print(customer_data.info())


print("\nDữ liệu bán hàng:")
print(sales_data.head())

print("\nDữ liệu khách hàng:")
print(customer_data.head())
```

# Xử lý trùng lặp nếu có

```python
print("Số lượng dòng dữ liệu bán hàng ban đầu: ", len(sales_data))
print("Số lượng dòng dữ liệu khách ban đầu: ", len(customer_data))
# Kiểm tra và loại bỏ các dòng trùng lặp trong dữ liệu bán hàng
sales_data = sales_data.drop_duplicates()
print("Số lượng dòng sau khi loại bỏ trùng lặp trong dữ liệu bán hàng:", len(sales_data))

# Kiểm tra và loại bỏ các dòng trùng lặp trong dữ liệu khách hàng
customer_data = customer_data.drop_duplicates()
print("Số lượng dòng sau khi loại bỏ trùng lặp trong dữ liệu khách hàng:", len(customer_data))
```

Số lượng dòng không đổi nghĩa là không có dữ liệu trùng lặp

# Xử lý giá trị thiếu nếu có 

```python
# Kiểm tra giá trị thiếu trong dữ liệu bán hàng
print("Giá trị thiếu trong dữ liệu bán hàng:")
print(sales_data.isnull().sum())

# Kiểm tra giá trị thiếu trong dữ liệu khách hàng
print("Giá trị thiếu trong dữ liệu khách hàng:")
print(customer_data.isnull().sum())


```

# Kiểm tra độ chính xác

```python
# Kiểm tra kiểu dữ liệu
print("Kiểu dữ liệu của dữ liệu bán hàng:")
print(sales_data.dtypes)

print("\nKiểu dữ liệu của dữ liệu khách hàng:")
print(customer_data.dtypes)

# Chuyển đổi kiểu dữ liệu 
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date']).dt.date

# Kiểm tra tính nhất quán (các giá trị trong cột Order)
print("Các giá trị duy nhất trong cột trạng thái bán hàng:")
print(sales_data['Order Status'].unique())


print(sales_data.describe())

```
